```python
import requests
from bs4 import BeautifulSoup

# Step 1: Define the target URL
url = "https://www.scrapethissite.com/pages/simple/"

# Step 2: Send an HTTP GET request to fetch the webpage content
response = requests.get(url)

# Step 3: Check if the request was successful
if response.status_code == 200:
    # Step 4: Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Step 5: Find all country container blocks (div tags with class 'country')
    countries = soup.find_all("div", class_="country")
    
    print(f"Successfully found {len(countries)} countries. Extracting data...\n")
    
    # Step 6: Loop through each container and extract specific information
    for country in countries[:10]:  # Limiting to the first 10 for clean output
        # Clean whitespaces using .strip()
        name = country.find("h3", class_="country-name").text.strip()
        capital = country.find("span", class_="country-capital").text.strip()
        population = country.find("span", class_="country-population").text.strip()
        area = country.find("span", class_="country-area").text.strip()
        
        # Display the collected details
        print(f"Country: {name}")
        print(f"  - Capital: {capital}")
        print(f"  - Population: {population}")
        print(f"  - Area: {area} sq km")
        print("-" * 30)
else:
    print(f"Failed to retrieve data. HTTP Status Code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup

def get_live_cricket_scores():
    # URL for NDTV Sports live cricket scores
    url = "https://sports.ndtv.com/cricket/live-scores"
    
    # Adding a User-Agent header helps bypass basic scraping blocks
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
    }
    
    print("Fetching live cricket data...")
    response = requests.get(url, headers=headers)
    
    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to retrieve data. Status code: {response.status_code}")
        return

    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Note: Inspect the exact class names on the live website, 
    # as they frequently undergo structural updates.
    match_cards = soup.find_all('div', class_='scr_match-block')
    
    if not match_cards:
        print("No live matches found or the website layout has changed.")
        return

    # Loop through all live matches found on the page
    for match in match_cards:
        # Extract the team names and scores
        teams = match.find_all('div', class_='scr_team-name')
        scores = match.find_all('div', class_='scr_team-score')
        
        # Ensure that both team and score exist before printing
        if teams and scores:
            team1 = teams[0].text.strip()
            score1 = scores[0].text.strip() if len(scores) > 0 else "Yet to Bat"
            
            team2 = teams[1].text.strip()
            score2 = scores[1].text.strip() if len(scores) > 1 else "Yet to Bat"
            
            print("---------------------------------")
            print(f"{team1}: {score1}")
            print(f"{team2}: {score2}")
            
            # Extract match status/result if available
            status = match.find('div', class_='scr_match-status')
            if status:
                print(f"Status: {status.text.strip()}")
        else:
            print("Match details currently unavailable.")

if __name__ == "__main__":
    get_live_cricket_scores()
```
