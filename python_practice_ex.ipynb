```python
%%markdown
## For loop Patterns .
```

```python
#Left-Aligned Right-Angled Triangle
rows = 5
for i in range(1, rows + 1):
    print("*" * i)
```

```python
#Right-Aligned Right-Angled Triangle
rows = 5
for i in range(1, rows + 1):
    # Spaces decrease as stars increase
    print(" " * (rows - i) + "*" * i)
```

```python
# Inverted Right-Angled Triangle
rows = 5
for i in range(rows, 0, -1):
    print("*" * i)
```

```python
#Full Pyramid
rows = 5
for i in range(1, rows + 1):
    # Print leading spaces followed by stars with spaces
    print(" " * (rows - i) + "* " * i)
```

```python
#Hollow Right-Angled Triangle
rows = 5
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        # Print star if it's the first col, last col, or last row
        if j == 1 or j == i or i == rows:
            print("*", end="")
        else:
            print(" ", end="")
    print()  # Move to next line
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        # Instance variables unique to each object
        self.owner = owner
        self.balance = balance

    # Instance method to read data
    def display_balance(self):
        return f"Account owner: {self.owner} | Balance: ${self.balance}"

    # Instance method to modify data
    def deposit(self, amount):
        self.balance += amount
        return f"Deposited ${amount}. New balance: ${self.balance}"


# 1. Create unique instances of the class
account1 = BankAccount("Alice", 1000)
account2 = BankAccount("Bob", 500)

# 2. Call instance methods on the objects
print(account1.display_balance())  # Output: Account owner: Alice | Balance: $1000
print(account2.display_balance())  # Output: Account owner: Bob | Balance: $500

# 3. Modify the internal state of account1
print(account1.deposit(250))       # Output: Deposited $250. New balance: $1250
print(account1.display_balance())  # Output: Account owner: Alice | Balance: $1250
```

```python
class Employee:       
    # Class variable (shared by all employees)
    company_name = "Hitachi"

    # Constructor to initialize instance variables
    def __init__(self, emp_name, emp_dept):               
        self.emp_name = emp_name
        self.emp_dept = emp_dept
        # pass

     # Instance Method
    def changes(self,new_comapny):             # new_company placeholder for changing to New company
        Employee.company_name = new_comapny
        
    # Instance Method
    # Method to print employee details
    def info(self):
        print(f"Employee {self.emp_name} works in {self.emp_dept} department of {self.company_name}")

    # static Method used in fixed interest or values to pass
    @staticmethod
    def addition(x,y):                    
        return x+y

# Creating an instance with the required arguments
emp1 = Employee("Amar", "Data Science")
emp3 = Employee("Kartik","cyber-security")

# Calling the method
emp1.info()
print(emp1.company_name)

emp1.changes("IBM")
print(emp1.company_name)
emp1.info()

emp3.company_name ="Hyosung"
emp3.info()

employ = emp1.addition(10,20)
print(employ)

```

```python
class MathUtils:
    @staticmethod
    def is_even(number):
        # Static methods do not use 'self' or 'cls'
        return number % 2 == 0

    @staticmethod
    def add(a, b):
        return a + b

# 1. Call static methods directly using the Class Name (Recommended)
result1 = MathUtils.is_even(4)
result2 = MathUtils.add(5, 10)

print(result1)  # Output: True
print(result2)  # Output: 15

# 2. Call static methods through a class instance (Also possible)
utils = MathUtils()
print(utils.is_even(7))  # Output: False
```

```python
# Multiple inheritance 

class country():
    def __init__(self, country_name, **kwargs):
        super().__init__(**kwargs)  # Passes remaining arguments up the chain
        self.country_name = country_name
        
    def country_info(self):
        print(f"country name is {self.country_name}")

class company():
    def __init__(self, com_name, **kwargs):
        super().__init__(**kwargs)  # Passes remaining arguments up the chain
        self.com_name = com_name
    
    def company_info(self):
        print(f"company name is {self.com_name}")

class employee(company, country):
    def __init__(self, emp_name, com_name, country_name):
        # super() automatically triggers initialization for both parent classes
        super().__init__(com_name=com_name, country_name=country_name)
        self.emp_name = emp_name
        
    def empl_info(self):
        print(f"employee name is {self.emp_name} working in {self.country_name} at {self.com_name}")

# Creating the object with quotes around strings
emp1 = employee('Suresh', 'xyz', 'Ind')
emp1.empl_info()

```

```python
#multileval inheritance
from datetime import datetime

# Get the current date and time
now = datetime.now()

class company():
    def __init__(self, com_name):
        self.com_name = com_name
    
    def company_info(self):
        print(f"company name is {self.com_name}")

    def tell_me(self):
        print("I am grand father")

class department(company):
    def __init__(self, dep_name,com_name):
        self.dep_name = dep_name
        self.com_name = com_name
    
    def dep_info(self):
        print(f"department name is {self.dep_name} of {self.com_name} ")

    def tell_me(self):
        print("I am father")

class employee(department):
    def __init__(self, emp_name,dep_name,com_name):
        self.emp_name = emp_name
        self.dep_name = dep_name
        self.com_name = com_name
    
    def emp_info(self):
        print(f"employee name is {self.emp_name} in department {self.dep_name} of {self.com_name} ")

    def show_today(self):
        print("Today's date and time:", now)

    def tell_me(self):
        print("I am grand son")

com1 = employee('Suresh', 'Networking', 'IBM')
com1.emp_info()
com1.show_today()
com1.tell_me()


        
```

```python
%%markdown
#### Polymorphism  one function with same name but different functionality we can use any where like fetching data from many sources
```

```python
# Base class (Optional, but good practice for structuring)
class DataFetcher:
    def fetch(self):
        pass

# Child classes implementing the shared 'fetch' method
class ApiFetch(DataFetcher):
    def fetch(self):
        print("fetching data from api")

class DatabaseFetch(DataFetcher):
    def fetch(self):
        print("fetching data from database")

class S3Fetch(DataFetcher):
    def fetch(self):
        print("fetching data from s3")

# This unified function demonstrates polymorphism.
# It handles ANY object that has a .fetch() method seamlessly.
def execute_fetch(fetcher_object):
    fetcher_object.fetch()

# 1. Create a list containing different types of fetcher objects
fetchers = [ApiFetch(), DatabaseFetch(), S3Fetch()]

# 2. Iterate through them uniformly
print("--- Polymorphism in Action ---")
for fetcher in fetchers:
    execute_fetch(fetcher)
```

```python
import time

# 1. Define the decorator function
def time_logger(original_function):
    # The wrapper function accepts any number of arguments (*args, **kwargs)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        
        # Execute the original function and save its result
        result = original_function(*args, **kwargs)
        
        end_time = time.time()
        duration = end_time - start_time
        print(f"[LOG] {original_function.__name__} took {duration:.4f} seconds to execute.")
        
        # Return the original function's result
        return result
    return wrapper

# 2. Apply the decorator using the @ symbol
@time_logger
def fetch_api_data(endpoint):
    print(f"Connecting to {endpoint}...")
    time.sleep(1.5)  # Simulate network delay
    return {"status": "success", "data": [1, 2, 3]}

@time_logger
def fetch_db_data(query):
    print(f"Running query: {query}...")
    time.sleep(0.5)  # Simulate database latency
    return ["row1", "row2"]

# 3. Call the decorated functions
print("--- Starting Data Fetching ---")
api_res = fetch_api_data("https://example.com")
db_res = fetch_db_data("SELECT * FROM products")

print(db_res)

```

```python
from datetime import date

class Person:
    # Class attribute shared by all instances
    species = "Homo sapiens"
    total_people = 0

    def __init__(self, name, age):
        # Instance attributes
        self.name = name
        self.age = age
        # Accessing class state to increment the population tracker
        Person.total_people += 1

    # Example 1: Class method to modify or view class-level state
    @classmethod
    def get_population_info(cls):
        # 'cls' refers to the Person class itself
        return f"Species: {cls.species}. Total population tracked: {cls.total_people}"

    # Example 2: Class method used as a Factory Method (Alternative Constructor)
    @classmethod
    def from_birth_year(cls, name, birth_year):
        # Calculate age and dynamically instantiate a new class object using 'cls'
        current_year = date.today().year
        calculated_age = current_year - birth_year
        return cls(name, calculated_age)


# --- How to Use It ---

# 1. Instantiate objects using the standard constructor
p1 = Person("Alice", 30)

# 2. Instantiate an object using the factory class method
p2 = Person.from_birth_year("Bob", 1996)

# 3. Call the class method directly on the Class
print(Person.get_population_info())
# Output: Species: Homo sapiens. Total population tracked: 2

# Verification of objects
print(f"{p1.name} is {p1.age} years old.")  # Output: Alice is 30 years old.
print(f"{p2.name} is {p2.age} years old.")  # Output: Bob is 30 years old.
```
